### Functions for cleaning text and sentiment analysis

```python
import re
positive_words=open('positiveword.txt','r').read().split('\n')
positive_words=positive_words[1:]
negative_words=open('negativeword.txt','r').read().split('\n')
def clean_text(text):
    text=text.lower()
    text=re.sub(r'[\*_:.,-]+',' ',text)
    text=re.sub(' +',' ',text)
    text=re.sub('\n',' ',text)
    text=re.sub(r'\d+$','',text)
    text=re.sub(r'\s+$','',text)
    return text  
def sentiment_analysis(text):
   
    pos_count=0
    neg_count=0
    
    for word in positive_words:
        if " " +word+" " in text:
            pos_count=+1
            
    for word in negative_words:
        if " "+word+" " in text:
            neg_count=+1        
    if pos_count > neg_count:
        if re.search(' not ', text):
            return 'Negative'
        else:
            return 'Positive'
    elif neg_count > pos_count:
        return 'Negative'
    
    return 'Neutral'
    
```

### Function for getting the product data

```python
from bs4 import BeautifulSoup

def data_dictionary(url):
    data ={}
    count=1
    name=''
    price=''
    reviews1=[]
    driver.get(url)
   
    try:
        data['Product_Name']=driver.find_element_by_class_name('pdp-mod-product-badge-title').text
    except:
        None
    try:    
        data['Product_Price']=driver.find_element_by_class_name('pdp-product-price').text
    except:
        None
        
    try:
       
        breaker=True
        pos_counter=0
        neg_counter=0
        neut_counter=0
        while True:            
            reviews=driver.find_element_by_class_name('mod-reviews')
            reviews=reviews.find_elements_by_class_name('item-content')
            for i in range(len(reviews)):
                if not re.search(r'Response from Store\s',reviews[i].text):
                    data['Review '+str(count)]=clean_text(reviews[i].text)
                    if (sentiment_analysis(data['Review '+str(count)])) =='Positive':
                        
                        pos_counter+=1
                    elif (sentiment_analysis(data['Review '+str(count)])) =='Negative':
                        neg_counter+=1
                    elif (sentiment_analysis(data['Review '+str(count)])) =='Neutral':
                        neut_counter+=1
                    count+=1;
            try:
                driver.find_element_by_class_name('next-btn next-btn-normal next-btn-medium next-pagination-item next').click()
            except:
                break
        
    except:
        None
  
    if count>1:
        for x in range(1,count):
            data['Sentiment '+str(x)]=sentiment_analysis(data['Review '+str(x)])    
        if pos_counter>neg_counter:
            if pos_counter>=neut_counter:
                data["Rating"]='Positive'
            else:
                data["Rating"]='Neutral'
        elif pos_counter<neg_counter:
            if neg_counter>=neut_counter:
                data['Rating']='Negative'
            else:
                data['Rating']='Neutral'
        else:
            data['Rating']='Neutral'
        
    return data  
```
